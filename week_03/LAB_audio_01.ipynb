```python
import torch
import torchaudio
import torchaudio.transforms as T # Resample을 위해 T라는 별칭으로 import
import matplotlib.pyplot as plt

# 1. 파일 목록 정의
file_names = [
    "095522039.m4a",
    "095522040.m4a",
    "095522041.m4a",
    "095522042.m4a"
]

# for 반복문으로 모든 파일 처리
for filename in file_names:
    print(f"--- 📂 현재 처리 중인 파일: {filename} ---")

    my_file_path = f"/content/{filename}"

    try:

        # 메타데이터 확인 및 오디오 불러오기
        metadata = torchaudio.info(my_file_path)
        waveform, sample_rate = torchaudio.load(my_file_path)
        print("Original Shape:", waveform.shape)
        print("Original Sample Rate:", sample_rate)

        # Waveform 그리기
        plt.figure()
        plt.title(f"Waveform of {filename}")
        plt.plot(waveform.t().numpy())
        plt.show()

        # Spectrogram 그리기
        specgram = T.Spectrogram()(waveform)
        plt.figure()
        plt.title(f"Spectrogram of {filename}")
        plt.imshow(specgram.log2()[0,:,:].numpy(), cmap='gray')
        plt.show()


        # Downsampling을 위한 새 샘플링 레이트 정의 (예: 8000Hz)
        new_sample_rate = 8000

        # Downsampler 생성 및 적용
        downsampler = T.Resample(orig_freq=sample_rate, new_freq=new_sample_rate, dtype=waveform.dtype)
        downsampled_waveform = downsampler(waveform)
        print(f"Shape after downsampling to {new_sample_rate}Hz: {downsampled_waveform.shape}")

        # Upsampler 생성 및 적용 (다시 원래 샘플링 레이트로 복원)
        upsampler = T.Resample(orig_freq=new_sample_rate, new_freq=sample_rate, dtype=downsampled_waveform.dtype)
        upsampled_waveform = upsampler(downsampled_waveform)
        print(f"Shape after upsampling back to {sample_rate}Hz: {upsampled_waveform.shape}")


    except Exception as e:
        print(f"'{filename}' 처리 중 오류 발생: {e}")

    print("-" * 40)
```

```python
import math
import timeit

import librosa
import matplotlib.colors as mcolors
import matplotlib.pyplot as plt
import pandas as pd
import resampy
from IPython.display import Audio

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)

DEFAULT_OFFSET = 201


def _get_log_freq(sample_rate, max_sweep_rate, offset):
    """Get freqs evenly spaced out in log-scale, between [0, max_sweep_rate // 2]

    offset is used to avoid negative infinity `log(offset + x)`.

    """
    start, stop = math.log(offset), math.log(offset + max_sweep_rate // 2)
    return torch.exp(torch.linspace(start, stop, sample_rate, dtype=torch.double)) - offset


def _get_inverse_log_freq(freq, sample_rate, offset):
    """Find the time where the given frequency is given by _get_log_freq"""
    half = sample_rate // 2
    return sample_rate * (math.log(1 + freq / offset) / math.log(1 + half / offset))


def _get_freq_ticks(sample_rate, offset, f_max):
    # Given the original sample rate used for generating the sweep,
    # find the x-axis value where the log-scale major frequency values fall in
    times, freq = [], []
    for exp in range(2, 5):
        for v in range(1, 10):
            f = v * 10**exp
            if f < sample_rate // 2:
                t = _get_inverse_log_freq(f, sample_rate, offset) / sample_rate
                times.append(t)
                freq.append(f)
    t_max = _get_inverse_log_freq(f_max, sample_rate, offset) / sample_rate
    times.append(t_max)
    freq.append(f_max)
    return times, freq


def get_sine_sweep(sample_rate, offset=DEFAULT_OFFSET):
    max_sweep_rate = sample_rate
    freq = _get_log_freq(sample_rate, max_sweep_rate, offset)
    delta = 2 * math.pi * freq / sample_rate
    cummulative = torch.cumsum(delta, dim=0)
    signal = torch.sin(cummulative).unsqueeze(dim=0)
    return signal


def plot_sweep(
    waveform,
    sample_rate,
    title,
    max_sweep_rate=48000,
    offset=DEFAULT_OFFSET,
):
    x_ticks = [100, 500, 1000, 5000, 10000, 20000, max_sweep_rate // 2]
    y_ticks = [1000, 5000, 10000, 20000, sample_rate // 2]

    time, freq = _get_freq_ticks(max_sweep_rate, offset, sample_rate // 2)
    freq_x = [f if f in x_ticks and f <= max_sweep_rate // 2 else None for f in freq]
    freq_y = [f for f in freq if f in y_ticks and 1000 <= f <= sample_rate // 2]

    figure, axis = plt.subplots(1, 1)
    _, _, _, cax = axis.specgram(waveform[0].numpy(), Fs=sample_rate)
    plt.xticks(time, freq_x)
    plt.yticks(freq_y, freq_y)
    axis.set_xlabel("Original Signal Frequency (Hz, log scale)")
    axis.set_ylabel("Waveform Frequency (Hz)")
    axis.xaxis.grid(True, alpha=0.67)
    axis.yaxis.grid(True, alpha=0.67)
    figure.suptitle(f"{title} (sample rate: {sample_rate} Hz)")
    plt.colorbar(cax)
```

```python
import torch
import torchaudio
import torchaudio.functional as F
import matplotlib.pyplot as plt

file_names = [
    "095522039.m4a",
    "095522040.m4a",
    "095522041.m4a",
    "095522042.m4a"
]

for filename in file_names:
    print(f"--- 📂 현재 처리 중인 파일: {filename} ---")

    my_file_path = f"/content/{filename}"

    try:
        waveform, sample_rate = torchaudio.load(my_file_path)
        print(f"Original Shape: {waveform.shape}, Sample Rate: {sample_rate}")

        downsample_rate = 8000
        downsampled_waveform = F.resample(waveform, sample_rate, downsample_rate)
        print(f"Downsampled to {downsample_rate}Hz Shape: {downsampled_waveform.shape}")

        upsample_rate = 10000
        upsampled_waveform = F.resample(downsampled_waveform, downsample_rate, upsample_rate)
        print(f"Upsampled to {upsample_rate}Hz Shape: {upsampled_waveform.shape}")

    except Exception as e:
        print(f"'{filename}' 처리 중 오류 발생: {e}")

    print("-" * 40)
```

```python
librosa_resampled_waveform = torch.from_numpy(
    librosa.resample(waveform.squeeze().numpy(), orig_sr=sample_rate, target_sr=resample_rate, res_type="kaiser_best")
).unsqueeze(0)
plot_sweep(librosa_resampled_waveform, resample_rate, title="Kaiser Window Best (librosa)")
```

```python
import torch
import torchaudio
import torchaudio.functional as F

file_names = [
    "095522039.m4a",
    "095522040.m4a",
    "095522041.m4a",
    "095522042.m4a"
]

# for 반복문으로 모든 파일을 한 번에 처리
for filename in file_names:
    print(f"--- 📂 현재 처리 중인 파일: {filename} ---")

    my_file_path = f"/content/{filename}"

    try:
        # 파일을 불러와 원본 파형(waveform)과 샘플링 레이트(sample_rate)를 얻음
        waveform, sample_rate = torchaudio.load(my_file_path)
        print(f"Original Shape: {waveform.shape}, Sample Rate: {sample_rate} Hz")

        # Downsampling 수행 (8000 Hz로 낮추기)
        downsample_rate = 8000
        downsampled_waveform = F.resample(waveform, orig_freq=sample_rate, new_freq=downsample_rate)
        print(f"Downsampled Shape: {downsampled_waveform.shape}, Sample Rate: {downsample_rate} Hz")

        # Upsampling 수행 (16000 Hz로 높이기)
        upsample_rate = 16000
        # 방금 다운샘플링된 파형을 입력으로 사용
        upsampled_waveform = F.resample(downsampled_waveform, orig_freq=downsample_rate, new_freq=upsample_rate)
        print(f"Upsampled Shape: {upsampled_waveform.shape}, Sample Rate: {upsample_rate} Hz")

    except Exception as e:
        print(f"'{filename}' 처리 중 오류 발생: {e}")

    print("-" * 40)
```
